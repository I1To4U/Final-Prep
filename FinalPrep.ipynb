```python
#Q3. การดำเนินการกับลิสต์ (List Operations)
#โจทย์: จัดการลิสต์ loads = [12.5, 9.8, 15.2, 11.0, 14.3]

loads = [12.5, 9.8, 15.2, 11.0, 14.3]

# 1. Print the highest load (using built-in max()) [cite: 102]
highest_load = max(loads)
print(f"1. Highest load: {highest_load}")

# 2. Loop and print all loads greater than 12 [cite: 103]
print("2. Loads > 12:")
for load in loads:
    if load > 12:
        print(f"   {load}")

# 3. Calculate and print the average load [cite: 104]
total_load = sum(loads)
average_load = total_load / len(loads)
# Average: 12.56
print(f"3. Average load: {average_load:.2f}")
```

```python
#Q4. การนับด้วยลูป (Counting With Loops)
#โจทย์: นับจำนวนค่าใน levels = [2.1, 2.4, 3.0, 4.7, 2.0, 5.1] ที่ตกอยู่ในช่วงต่างๆ

levels = [2.1, 2.4, 3.0, 4.7, 2.0, 5.1]

low_count = 0     # readings below 2.5 [cite: 109]
normal_count = 0  # readings between 2.5 and 4.5 [cite: 110]
high_count = 0    # readings above 4.5 [cite: 111]

for level in levels:
    if level < 2.5:
        low_count += 1
    elif level <= 4.5:  # Implies 2.5 <= level <= 4.5
        normal_count += 1
    else:               # Implies level > 4.5
        high_count += 1

# Print the counts [cite: 112-115]
print(f"Low: {low_count}")      # Should be 3 (2.1, 2.4, 2.0)
print(f"Normal: {normal_count}") # Should be 1 (3.0)
print(f"High: {high_count}")     # Should be 2 (4.7, 5.1)
```

```python
#Q5. ฟังก์ชันที่มีค่าส่งคืน (Function With Return Value)
#โจทย์: เขียนฟังก์ชัน calc_area(length, width) เพื่อคำนวณและส่งคืนพื้นที่

# Function Definition [cite: 117-121]
def calc_area(length, width):
    """Computes the area of a rectangle."""
    area = length * width
    return area

# Main Program [cite: 122-124]
footing_length = 4
footing_width = 3

# Call the function and store the output [cite: 122]
footing_area = calc_area(footing_length, footing_width)

# Print the result
print(f"Area = {footing_area}")
```

```python
#Q6. ฟังก์ชันที่ส่งคืนสองค่า (Function Returning Two Values)
#โจทย์: เขียนฟังก์ชันที่หาและส่งคืนทั้งค่าต่ำสุดและสูงสุดจากลิสต์

# Function Definition [cite: 126-129]
def find_min_max(data_list):
    """Returns the minimum and maximum values from a list."""
    # Using built-in functions for simplicity
    min_val = min(data_list)
    max_val = max(data_list)
    # Python returns multiple values as a tuple
    return min_val, max_val

# Main Program [cite: 130-132]
data = [12, 7, 19, 5, 14]

# Call the function and unpack the two returned values [cite: 131-132]
# The variables must be in the same order as the return statement
minimum_result, maximum_result = find_min_max(data)

print(f"List: {data}")
print(f"Minimum value: {minimum_result}") # 5
print(f"Maximum value: {maximum_result}") # 19
```

```python
#Q7. คลาสที่มีแอตทริบิวต์ส่วนตัว (Class With Private Attribute)
#โจทย์: สร้างคลาส Sensor พร้อมแอตทริบิวต์ส่วนตัว _limit, เมท็อด is_over(), และเมท็อด getter get_limit()

class Sensor:
    # Constructor [cite: 135, 57]
    def __init__(self, sensor_id, limit):
        self.id = sensor_id
        self.readings = []
        # Private attribute convention is to use a leading underscore [cite: 135]
        self._limit = limit 

    # Method to check if the last reading exceeds the limit [cite: 136]
    def is_over(self):
        if not self.readings:
            return False
        # Accesses the last reading using negative indexing
        last_reading = self.readings[-1]
        return last_reading > self._limit

    # Getter method to safely access the private limit [cite: 137, 65]
    def get_limit(self):
        return self._limit

# Create an object and test methods [cite: 138]
# Create sensor with ID 'T1' and limit 25.0
temp_sensor = Sensor(sensor_id='T1', limit=25.0)

# Add some readings
temp_sensor.readings.append(22.5)
temp_sensor.readings.append(26.1) # This one is OVER the limit

# Test the get_limit method [cite: 138]
print(f"Sensor ID: {temp_sensor.id}, Limit: {temp_sensor.get_limit()}")

# Test the is_over method [cite: 138]
print(f"Last reading: {temp_sensor.readings[-1]}")
print(f"Is over limit? {temp_sensor.is_over()}")

# Add a reading that is NOT over the limit
temp_sensor.readings.append(24.9)
print(f"\nNew last reading: {temp_sensor.readings[-1]}")
print(f"Is over limit? {temp_sensor.is_over()}")
```
